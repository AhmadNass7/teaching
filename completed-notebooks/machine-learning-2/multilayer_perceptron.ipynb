# Multilayer Perceptron
Load the `mnist` dataset. Split it into training and test sets. Train and test a multilayer perceptron model using scikit-learn. Check the documentation to identify the most important hyperparameters, attributes, and methods of the model. Use them in practice.

## Importing Modules

```python
import pandas as pd
import sklearn.metrics
import sklearn.linear_model
import sklearn.model_selection
import sklearn.neural_network
import plotly.express as px
```

## Loading the Dataset

```python
df = pd.read_csv("../../datasets/mnist.csv")
df = df.set_index("id")
df.head(3)
```

## Splitting the Data into Training and Test

```python
x = df.drop(["class"], axis=1)
y = df["class"]
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y)
```

## Model Selection and Hyperparameter Tuning

```python
parameters_grid = {
    "solver": ["lbfgs", "sgd", "adam"],
    "batch_size": [10, 20, 50, 100, 200],
    "max_iter": [200, 400],
    "hidden_layer_sizes": [(50), (100), (50, 50), (50, 100), (100, 50), (100, 100)]
}
model_1 = sklearn.model_selection.GridSearchCV(sklearn.neural_network.MLPClassifier(), 
                                               parameters_grid, scoring="accuracy", cv=5, n_jobs=-1)
model_1.fit(x_train, y_train)
print("Accuracy of best Random Forest classfier = {:.2f}".format(model_1.best_score_))
print("Best found hyperparameters of Random Forest classfier = {}".format(model_1.best_params_))
```

## Testing the Trained Model

```python
y_predicted = model_1.predict(x_test)
accuracy = sklearn.metrics.accuracy_score(y_test, y_predicted)
cm = sklearn.metrics.confusion_matrix(y_test, y_predicted)
precision, recall, f1, support = sklearn.metrics.precision_recall_fscore_support(y_test, y_predicted)

print("Accuracy =", accuracy)
print("Precision =", precision)
print("Recall =", recall)
print("F1-Score =", f1)
print("Confusion Matrix:\n", cm)
```
